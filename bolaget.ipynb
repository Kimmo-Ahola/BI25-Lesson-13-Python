```python
import pandas as pd

# eftersom filen har systembolagets logga på första raderna kör vi header
# för att skippa dessa rader
df = pd.read_excel("Artikellistan2024.xlsx", header=4)
```

```python
# Vi kan med denna rad inspektera vilka rader som har flest null
df.isnull().sum()

# ekologiskt och etiskt har många null-värden
# vi tar bort dem i nästa steg
```

```python
# Vi droppar dessa kolumner.
# eftersom det är flera kolumner blir det en lista
df = df.drop(["Ekologisk", "Etiskt"], axis=1)
df # denna printar ut resultatet
```

```python
# Jag vill ändra namn och enhet på volym i ml
df = df.rename(columns={"Volym i ml" : "Volym i liter"})

# räknar om så att ml blir till liter genom att dividera med 1000
df["Volym i liter"] = df["Volym i liter"] / 1000
df # printa ut resultatet
```

```python
# Räkna ut antal sålda enheter
df["Enheter sålda"] = df["Försäljning i liter"] / df["Volym i liter"]
df # printa ut resultatet
```

```python
# Jag vill veta vilka produkter med Sverige som ursprung som har sålt mest

# Filtrera bort allt som inte har ursprung Sverige
df_swedish = df[df["Ursprung"] == "Sverige"]

# Sortera per enheter sålda i fallande ordning
# fallande ordning = ascending = False
df_swedish = df_swedish.sort_values(by="Enheter sålda", ascending=False)

# Vi kör en slice för att få topp 15
df_swedish = df_swedish[:15] # top 15 most sold. Uses slicing
df_swedish # printa ut resultatet
```

```python
population_over_15 = 8.5e6 # 8.5 million over 15 in swedish
# how many units per person over 15?
# vuxen = over 15 in this case
# Vi vill få ut antal sålda enheter per vuxen per år (2024)
df_swedish["Enheter per vuxen"] = df_swedish["Enheter sålda"] / population_over_15
df_swedish # printa ut resultatet
```

```python
import matplotlib.pyplot as plt

# vi filtrerar ut endast de två kolumner vi är intresserade av
df_swedish = df_swedish[["Kvittonamn", "Enheter per vuxen"]]

# stapelgraf med kvittonamn och enheter per vuxen
plt.bar(df_swedish["Kvittonamn"], df_swedish["Enheter per vuxen"])

# Vi sätter text på x-axeln
plt.xlabel("Kvittonamn")
# Vi sätter text på y-axeln
plt.ylabel("Antal sålda enheter")

# vi sätter titel på vår graf
plt.title("Enheter sålda per vuxen år 2024")

# Vi roterar texten med 90 grader så att den inte överlappar
# kvittonamnen är ganska långa
plt.xticks(rotation=90)

# Superviktig! Vi vill visa grafen
plt.show()
```

```python
# Som sista steg så sparar vi vårt nya dataset
# Spara alltid som en ny fil. 
# Skriv aldrig över originalet
df_swedish.to_excel("Artikellistan_2024_topp_15.xlsx", index=False)
```
