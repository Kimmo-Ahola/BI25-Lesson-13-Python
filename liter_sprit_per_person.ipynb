```python
import pandas as pd

df = pd.read_excel("Artikellistan2024.xlsx", header=4)
```

```python
# Vi extraherar siffervärdet för alkoholprocenten
# från kvittonamn med regex
alc_regex = r"\b(\d{1,2}(?:[.,]\d{1,2})?\s?)%"
df["Alkoholhalt"] = df["Kvittonamn"].str.extract(alc_regex)

# eftersom vi i Sverige har , istället för .
# omvandlar vi detta med str.replace så att vi kan
# göra en float av värdet

df["Alkoholhalt"] = df["Alkoholhalt"].str.replace(',', '.')
df["Alkoholhalt"] = df["Alkoholhalt"].astype(float)
df
```

```python
# Vi vill filtrera bort allt som saknar alkoholhalt

df = df[df["Alkoholhalt"] > 0]
df
```

```python
# ändra volym i ml till volym i liter för enklare beräkning
df = df.rename(columns={"Volym i ml": "Volym i liter"})
df["Volym i liter"] = df["Volym i liter"] / 1000
df
```

```python
df["Mängd i 40 %"] = (df["Försäljning i liter"] * df["Alkoholhalt"] / 100) / 0.4
df
```

```python
population_over_15 = 8.5e6
sum_of_liters = df["Mängd i 40 %"].sum()
res = sum_of_liters / population_over_15
res

# Officiella svaret är ca 8,4 liter 40 %-ig sprit per person
# ganska nära!
```

```python
df["Alkohol per krona"] = (df["Volym i liter"] * df["Alkoholhalt"] / 100) / df["Aktuellt pris"]
df = df.sort_values("Alkohol per krona", ascending=False)
df
```

```python
# Mest alkohol för pengarna
df[["Kvittonamn", "Alkohol per krona", "Aktuellt pris"]].head(10)
```

```python
# Minst alkohol för pengarna
df[["Kvittonamn", "Alkohol per krona", "Aktuellt pris"]].tail(10)
```

```python
import matplotlib.pyplot as plt

# filter out low selling products
df_filtered = df[df["Försäljning i liter"] > 0.02e7]
x = df_filtered["Alkohol per krona"]
y = df_filtered["Försäljning i liter"]

plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.6)

plt.xlabel("Alkohol per krona")
plt.ylabel("Försäljning i liter")
plt.title("Alkohol per krona vs Försäljning i liter")

plt.show()
```

```python
import matplotlib.pyplot as plt

top_n = 10

# sort indices by sales
sorted_idx = df_filtered["Försäljning i liter"]\
    .sort_values(ascending=False)\
    .index

# create groups of 10
top_groups = []
for i in range(0, 40, top_n):
    top_groups.append(sorted_idx[i:i+top_n])

colors = ["red", "orange", "green", "purple"]

plt.figure(figsize=(8,6))

plt.scatter(
    df_filtered["Alkohol per krona"],
    df_filtered["Försäljning i liter"],
    color="lightblue",
    alpha=0.3
)

for i, group in enumerate(top_groups):
    plt.scatter(
        df_filtered.loc[group, "Alkohol per krona"],
        df_filtered.loc[group, "Försäljning i liter"],
        color=colors[i],
        s=50,
        label=f"Top {i*10+1}–{(i+1)*10}"
    )

plt.xlabel("Alkohol per krona")
plt.ylabel("Försäljning i liter")
plt.title("Alkohol per krona vs Försäljning i liter")
plt.legend()

plt.show()
```
