```python
import pandas as pd

df = pd.read_excel("Artikellistan2024.xlsx", header=4)
```

```python
# Vi extraherar siffervärdet för alkoholprocenten
# från kvittonamn med regex
alc_regex = r"\b(\d{1,2}(?:[.,]\d{1,2})?\s?)%"
df["Alkoholhalt"] = df["Kvittonamn"].str.extract(alc_regex)

# eftersom vi i Sverige har , istället för .
# omvandlar vi detta med str.replace så att vi kan
# göra en float av värdet

df["Alkoholhalt"] = df["Alkoholhalt"].str.replace(',', '.')
df["Alkoholhalt"] = df["Alkoholhalt"].astype(float)
df
```

```python
# Vi vill filtrera bort allt som saknar alkoholhalt

df = df[df["Alkoholhalt"] > 0]
df
```

```python
# ändra volym i ml till volym i liter för enklare beräkning
df = df.rename(columns={"Volym i ml": "Volym i liter"})
df["Volym i liter"] = df["Volym i liter"] / 1000
df
```

```python
df["Mängd i 40 %"] = (df["Försäljning i liter"] * df["Alkoholhalt"] / 100) / 0.4
df
```

```python
population_over_15 = 8.5e6
sum_of_liters = df["Mängd i 40 %"].sum()
res = sum_of_liters / population_over_15
res

# Officiella svaret är ca 8,4 liter 40 %-ig sprit per person
# ganska nära!
```

```python
df["Alkohol per krona"] = (df["Volym i liter"] * df["Alkoholhalt"] / 100) / df["Aktuellt pris"]
df = df.sort_values("Alkohol per krona", ascending=False)
df
```
